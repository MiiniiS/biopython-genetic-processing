Sequence complement and reverse complement

```python
from IPython.core.interactiveshell import validate
from Bio.Seq import Seq
DNA_Seq=Seq("ATCGGGGTCATTAAAAT")
DNA_Seq.complement()
DNA_Seq.reverse_complement()
print("DNA Complement Stand",DNA_Seq.complement)
print("DNA reverse Complement Stand",DNA_Seq.reverse_complement())
```

Transcription (DNA -> RNA)

```python
DNA = Seq("ATCGGGGTCATTAAAAT")
Template_DNA= DNA.reverse_complement()
MSG_rna=DNA.transcribe()
MSG_rna.back_transcribe()
print("Meg RNA",MSG_rna)
print("Meg RNA Back Transcribe",MSG_rna.back_transcribe())
```
